```python
import pandas as pd
import csv
```

```python
df_netflix = pd.read_csv('netflix_titles.csv')
```

```python
df_netflix
```

```python
replace_map = {
    'United States': 'UnitedStates',
    'United Kingdom': 'UnitedKingdom',
    'United Arab Emirates': 'UnitedArabEmirates',
    'Hong Kong': 'HongKong',
    'New Zealand': 'NewZealand',
    'Saudi Arabia': 'SaudiArabia',
    'Czech Republic': 'CzechRepublic',
    'South Korea': 'SouthKorea',
    'West Germany': 'WestGermany',
    'East Germany': 'EastGermany',
    'Cayman Islands': 'CaymanIslands',
    'South Africa': 'SouthAfrica',
    'Soviet Union': 'SovietUnion',
    'Sri Lanka' : 'SriLanka',
    'Vatican City' : 'VaticanCity',
    'Dominican Republic' : 'DominicanRepublic'
}

df_netflix['country'] = df_netflix['country'].apply(lambda x: ', '.join(replace_map.get(val.strip(), val) for val in x.split(',')) if isinstance(x, str) else x)
single_column = df_netflix['country'].str.replace(',', ' ')
single_column = single_column.fillna('NoData')
```

```python
single_column
```

```python
single_column.to_csv('country.csv', index=False, quoting=csv.QUOTE_NONE, header=False)
```
